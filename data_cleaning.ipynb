```python
from dask import dataframe as df
import pandas as pd
import functools as ft
```

```python
dask_df1 = df.read_csv('title_akas.tsv', sep ='\t', dtype={'isOriginalTitle': 'object'})
dask_df2 = df.read_csv('title_basics.tsv', sep ='\t', dtype={'runtimeMinutes': 'object',
       'startYear': 'object', 'isAdult': 'object'})
dask_df3 = df.read_csv('title_rating.tsv', sep ='\t')
```

```python
dask_df1 = dask_df1.rename(columns={'titleId':'tconst'})
dask_df1.head()
```

```python
dask_df2.head(1000)
```

```python
dask_df3.head()
```

```python
dask_df = dask_df1.merge(dask_df2, how = 'left', on = 'tconst')
```

```python
dff = dask_df.merge(dask_df3, how = 'left', on = 'tconst')
```

```python
dff.head()
```

```python
dff.columns
```

```python
dff = dff.drop(['ordering', 'region', 'attributes', 'isOriginalTitle', 'types', 'primaryTitle',
       'originalTitle', 'isAdult', 'startYear', 'endYear', 'runtimeMinutes'], axis = 1)
```

```python
dff.head()
```

```python
dff1 = dff[dff['titleType']=='movie']
```
